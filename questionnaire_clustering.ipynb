```python
from pandas import *
from sklearn import cluster
import warnings
warnings.filterwarnings('ignore')
```

```python
########### IMPORT DATA ###########
data = read_csv("generated_results/data_sample_15.csv",sep=";")
columns_to_keep = ["SKILL_C_CC","SKILL_I_CC","SKILL_A_CC"]
df_data = DataFrame(data=data)
df_data.rename(columns={df_data.columns[0]:"User_ID"},inplace=True)


set_option('display.max_rows',None)
column_entity = ["CC","DS","DF", "EE", "TB"]
human_factors = ["Skill","Experience"]
security_aspects_fn = ["Confidentiality","Integrity","Authenticity"]

security_aspects = ["C","I","A"]
df_total_users = df_data[columns_to_keep]

df_total_users
```

```python
###### K MEANS FIRST PART 3 GROUPS IN TOTAL ######
nb_grp = 3

k_means = cluster.KMeans(n_clusters=nb_grp)
k_means.fit_predict(df_total_users)

labels = k_means.labels_

results = DataFrame([df_total_users.index,labels]).T
results.columns = ["User_ID","Group"]
results


df_total_users_groups = concat([df_total_users,results],axis=1)
df_total_users_groups.sort_values(by="User_ID")
```
